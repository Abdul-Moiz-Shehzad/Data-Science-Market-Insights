```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data Science Jobs 2024.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
print("before",len(df))
df=df.drop_duplicates()
print("after",len(df))
```

```python
df=df.loc[df["Salary Estimate"]!='-1']
len(df)
```

* Salary

```python
df["Salary Estimate"].unique()
```

```python
df["Employer est."]=df["Salary Estimate"].apply(lambda x: 1 if x.find('Employer est.')!=-1 else 0)
```

```python
df["Per hour"]=df["Salary Estimate"].apply(lambda x: 1 if x.find('Per Hour')!=-1 else 0)
```

```python
EstRemoved=df["Salary Estimate"].apply(lambda x: x.split(' (')[0])
```

```python
removed_dk=EstRemoved.apply(lambda x: x.replace('$','').replace('K',''))
df["Min Salary"]=removed_dk.apply(lambda x: float(x.split(' - ')[0]))
removed_dk
```

```python
removed_perhour=removed_dk.apply(lambda x: x.replace(' Per Hour',''))
removed_perhour
```

```python
df["Max Salary"]=removed_perhour.apply(lambda x: float(x.split(' - ')[1]) if x.find(' - ')!=-1 else float(x))
```

```python
df["Avg Salary"]=(df["Min Salary"]+df["Max Salary"])/2
```

* Job Title

```python
#df["Job Title"][df["Job Title"].str.contains("")]
df["Job Title"].unique()
```

```python
def JobTitleClean(title):
    if ("data scientist" in title.lower() or "science" in title.lower()) and (" ai " not in title.lower()):
        return "data scientist"
    elif ("data engineer" in title.lower() or "engineer" in title.lower()) and (" ai " not in title.lower()):
        return "data engineer"
    elif ("data analytics" in title.lower() or "analytics" in title.lower() or "analyst" in title.lower() or "insight" in title.lower()) and (" ai " not in title.lower()):
        return "data analyst"
    elif " ai " in title.lower() or "artifical intelligence" in title.lower() or "ai " in title.lower():
        return "artifical intelligence"
    elif "machine learning" in title.lower() or "/ ml" in title.lower():
        return "machine learning"
    elif "instructor" in title.lower() or "Lecturer" in title.lower() or "professor" in title.lower() or "class" in title.lower() or "supervisor" in title.lower() or "director" in title.lower():
        return "teaching"
    elif "statistician" in title.lower():
        return "statistician" 
    elif "deep" in title.lower():
        return "deep learning"
    elif "marketing" in title.lower():
        return "marketing related"
    elif "applied" in title.lower():
        return "applied scientist"
    elif "research" in title.lower():
        return "research"
    elif "business" in title.lower():
        return "business development"
    else:
        return "unknown"
df["Refined Title"]=df["Job Title"].apply(JobTitleClean)
```

```python
df=df[df["Refined Title"]!="unknown"]
```

```python
def seniority(title):
    if 'sr' in title.lower() or "expert" in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'principal' in title.lower() :
            return 'senior'
    elif 'president' in title.lower() or 'head' in title.lower() or"supervisor" in title.lower() or 'lead' in title.lower() or "associate" in title.lower() or "vp" in title.lower() :
        return 'head'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'jr' in title.lower() or "junior" in title.lower():
        return 'jr'
    elif "staff" in title.lower():
        return "staff"
    else:
        return 'not applicable'
seniorityOfJobs=df["Job Title"].apply(seniority)
```

```python
seniorityOfJobs.unique()
df["Seniority"]=seniorityOfJobs
```
